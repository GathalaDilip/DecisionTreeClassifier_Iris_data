```python
 # Import necessary libraries
import pandas as pd  # For data manipulation
from sklearn.model_selection import train_test_split  # For splitting the data
from sklearn.tree import DecisionTreeClassifier, plot_tree  # For the decision tree classifier and plotting
from sklearn import metrics  # For evaluating the model
import matplotlib.pyplot as plt  # For plotting
```

```python
# Mount Google Drive to access files (specific to Google Colab environment)
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the Iris dataset from the CSV file in Google Drive
df = pd.read_csv('/content/drive/MyDrive/Iris_data.csv')
```

```python
# Display the first 5 rows of the dataset to understand its structure
df.head()
```

```python
# Check the shape of the dataset (number of rows and columns)
df.shape
```

```python
# Check for any missing values in the dataset
df.isnull().sum()
```

```python
# Convert the species names to numerical values for easier processing
# Iris-setosa -> 0, Iris-versicolor -> 1, Iris-virginica -> 2
df['Species'] = df['Species'].map({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2})
```

```python
# Display the first 5 rows again to see the changes
df.head()
```

```python
# Split the Data into Features (X) and Labels (y)
# Features are the attributes used to make the predictions (excluding 'Id' and 'Species')
X = df.drop(['Id', 'Species'], axis=1)
# Labels are the actual outcomes we want to predict
y = df['Species']
```

```python
# Split the data into Training and Testing sets
# Training set is used to train the model, Testing set is used to evaluate it
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train the Decision Tree Classifier
clf = DecisionTreeClassifier()  # Create an instance of the Decision Tree Classifier
clf = clf.fit(X_train, y_train)  # Train the classifier with the training data
```

```python
# Use the trained classifier to make predictions on the testing data
y_pred = clf.predict(X_test)
```

```python
# Calculate and print the accuracy of the model
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
# Plot the decision tree to visualize it
plt.figure(figsize=(20,10))  # Set the size of the plot
plot_tree(clf, filled=True, feature_names=X.columns, class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.show()
```
